```python
# Project Introduction and Goals

print("üìö What the Project is About:")
print("This project analyzes the Netflix Movies dataset to explore trends in movie ratings, release patterns, and genre distribution on Netflix. It involves data cleaning, visualization, and deriving insights from the data.")

print("\nüéØ What the Goal of the Analysis Was:")
print("The goal of this analysis was to understand Netflix's movie content strategy by identifying the most common ratings, peak release periods, and popular genres among the movies available on the platform.")

print("\nüîç Why I Chose the Dataset:")
print("I chose the Netflix Movies dataset because Netflix is one of the largest streaming platforms globally, and analyzing its movie offerings can provide valuable insights into audience preferences, content trends, and entertainment industry shifts.")

#SETUP
# Step 1: Import Pandas
import pandas as pd

# Step 2: Load the dataset
df = pd.read_csv('netflix_titles.csv.zip')  # Make sure the file is uploaded to Colab

# Step 3: Explore the data
print("First 5 rows:")
print(df.head())

print("\nData Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

# Step 4: Basic Data Cleaning
# Keep only 'Movie' entries
df = df[df['type'] == 'Movie']

# Drop rows where 'rating' is missing
df = df.dropna(subset=['rating'])

# Optional: Keep only important columns
df = df[['title', 'rating', 'listed_in', 'release_year', 'duration']]

# Check if cleaning worked
print("\nCleaned Data:")
print(df.head())

# (Optional) Visual check for missing values (you can comment it out if you want)
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False)
plt.title('Missing Values Heatmap')
plt.show()

#IMPORT SEABORN AND MATPLOTIB
# Already imported at the end of setup (Seaborn and Matplotlib)
# If not, you can import here again

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

# 1. Plot: Number of Movies by Rating
print("The plot shows the sequence of movies by their rating")
print("The plot gives clear picture about movies through their ratings")
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index, palette='muted')
plt.title('Number of Movies by Rating')
plt.xticks(rotation=45)
plt.show()

# 2. Plot: Movies Released Per Year
print("The plot shows movies which are realeased every year")
print("The plot dives into the releaes")
plt.figure(figsize=(12,6))
df['release_year'].value_counts().sort_index().plot(kind='line', color='green')
plt.title('Movies Released Over the Years')
plt.grid(True)
plt.show()

# 3. (Optional) Plot: Top 10 Genres
print("this plot gives details about the genres that are mostly viewed")
print("It also gives insights about the top 10")
plt.figure(figsize=(12,6))
df['listed_in'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Movie Genres')
plt.xticks(rotation=45)
plt.show()

# 4. (Optional) Plot: Pie Chart of Ratings
plt.figure(figsize=(8,8))
df['rating'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Ratings Distribution')
plt.show()

# Insights from Netflix Movies Dataset Analysis

print("üîπ 1. The most common movie rating on Netflix is 'TV-MA', which indicates content for mature audiences.")
print("üîπ 2. The number of movie releases peaked around the year 2017, suggesting Netflix had a major content expansion during that time.")
print("üîπ 3. Drama is the most popular genre among Netflix movies, followed by Comedies and Documentaries.")
print("üîπ 4. The majority of Netflix movies have a release date after 2010, showing a focus on modern content.")
print("üîπ 5. A significant portion of movies are targeted towards adult audiences rather than children or family categories.")


# Sample data insights based on your analysis
most_common_rating = df['rating'].mode()[0]  # Find the most frequent rating
num_movies_per_year = df.groupby('release_year').size().idxmax()  # Year with most movie releases
# Change 'genre' to 'listed_in' to analyze genres based on available column
top_genres = df['listed_in'].value_counts().head(3)  # Top 3 genres

# Conclusion generated dynamically
print("Conclusion:")
print(f"\n1. The most common rating on Netflix is '{most_common_rating}', indicating Netflix primarily targets mature audiences.")
print(f"\n2. The year with the highest number of movie releases is {num_movies_per_year}.")
print(f"\n3. The top 3 genres are:")
for genre, count in top_genres.items():
    print(f"    - {genre}: {count} movies")

# General Summary
print("\nOverall, Netflix is focused on catering to mature audiences with a diverse range of popular genres.")


```
